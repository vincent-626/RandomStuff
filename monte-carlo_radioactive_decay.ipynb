```python
import numpy as np
from matplotlib import pyplot as plt

# Define values
t_half = 10
n0 = 1000
n1 = 0
lamb = np.log(2)/t_half

# Create figure and axes
fig, ax = plt.subplots()

# Theoretical curve
t = np.linspace(0, 50, 1000)
x = np.linspace(0, 5, 1000)
n = n0*np.exp(-lamb*t)

ax.plot(x, n, label='Theory')

# Remaining atom curve
dt = 50/999
p = 1 - 2**(-dt/t_half)
y1 = [1000]
y2 = [0]

for i in range(len(t)-1):
    for j in range(n0):
        r = np.random.uniform()
        if r < p:
            n0 -= 1
            n1 += 1
    y1.append(n0)
    y2.append(n1)

ax.plot(x, y1, label='Remaining atoms, Monte-Carlo')

# Decayed curve
ax.plot(x, y2, label='Decayed atoms, Monte-Carlo')

# Plotting styles
ax.set_xlabel('Half lives')
ax.set_ylabel('Number of atoms')
plt.title('Simple Monte-Carlo Decay Model')
plt.legend()
plt.show()
```
